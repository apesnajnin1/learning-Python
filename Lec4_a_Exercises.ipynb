```python
balance = float(input('Enter a balance = '))
if balance < 0:
  print ('False')
else:
  withdraw_amount = float(input('Enter a amount = '))
  if withdraw_amount > balance:
    print ('Insufficient Balance')
  else:
    amount = withdraw_amount
    balance = balance - amount
    print ('Remaining Balance: {}'.format(balance))
```

```python
# Identify the Traingle type
a = int(input('Enter a side = '))
b = int(input('Enter a side = '))
c = int(input('Enter a side = '))
if a == b == c:
  print ('Equilateral Triangle')
elif a == b or b == c or c == a:
  print ('Isosceles Triangle')
else:
  print ('Scalene Triangle')

# Test the validity of Triangle
if a + b > c and b + c > a and c + a > b:
  print ('Valid Triangle')
else:
  print ('Invalid Triangle')
```
