```python
#Question2:
#ii. Identify the average study hours per day among
#high-performing students ( e.g. those scoring above 85%)

import pandas as pd

# Load the std>performance.csv file
df = pd.read_csv('std_performance.csv')

std_high_performance = df[df['exam_score'] > 85]
print(std_high_performance)
std_high_performance.to_csv('std_high_performance.csv')

# Calculate average study hours per day among high-performance students
average = df['study_hours_per_day'].mean()

print(f"The average study hours per day among high_performing students is: {average}")
```

```python
#Question3:
#i. Find the Percentage of students who
#regularly attend classes ( e.g., those attended above 80%)
import pandas as pd
df = pd.read_csv('student_habits_performance.csv')

Regular_students = df[['attendance_percentage', 'exam_score']]
print(Regular_students)
Regular_students.to_csv('Regular_students.csv')


df1 = pd.read_csv('Regular_students.csv')

std_high_attendance = df1[df1['attendance_percentage'] > 80]
print(std_high_attendance)
std_high_attendance.to_csv('std_high_attendance.csv')


df2 = pd.read_csv('std_high_attendance.csv')
total_students = len(df)
total_students_high_attendance = len(df2)
percentage_of_students = (total_students_high_attendance/total_students) * 100
print(f"The percentage of students who regularly attend classes is: {percentage_of_students}%")
```

```python
#Question4:
#i. The average Mobile Phone Usage ( social media hours)

std_mobile_phone_user = df[['social_media_hours','exam_score']]
print('Stuents list who use mobiles:')
print(std_mobile_phone_user)
std_mobile_phone_user.to_csv('std_mobile_phone_user.csv')


average = df['social_media_hours'].mean()
print(f"The average mobile phone usage is: {average}")

print('\n')
print('\n')
print('Studentslist who are high mobile users:')
High_average = std_mobile_phone_user[std_mobile_phone_user['social_media_hours'] > average]
print(High_average)
length_of_High_average = len(High_average)
print(f"The length of High_average is: {length_of_High_average}")
High_average.to_csv('High_average.csv')

print('\n')
print('\n')
print('Studentslist who are high mobile users:')
Low_average = std_mobile_phone_user[std_mobile_phone_user['social_media_hours'] < average]
print(Low_average)
length_of_Low_average = len(Low_average)
print(f"The length of Low_average is: {length_of_Low_average}")
Low_average.to_csv('Low_average.csv')

```
