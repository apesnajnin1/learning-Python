```python
F = int(input("Enter the temperature in Farenheit"))
print(f"F = {F}")
```

```python
C = (F - 32)*5/9
print('C = {:.2f}'.format(C))
```

```python
if C < 0:
  print ('Freezing')
  if C < -30:
    print ('Extremely cold!')
else:
  print ('Above Freezing')
  if C >= 40:
    print ('Extremely Hot!')
```

```python
import math

m = int(input("Put a number: "))
prime = None
if m == 2:
  prime = True
elif m>2 and m % 2 == 0:
  prime = False
else:
    i = 3
    while (i <= int( math.sqrt(m)+1)):
      if m % i == 0:
        prime = False
        break
      else:
        i += 2

if i > int(math.sqrt(m)):
  prime = True
  if prime:
    print(f"{m} is a prime number")
  else:
    print(f"{m} is not a prime number")
```
